## Advanced Matplotlib: Twin Axes, Styling, and Exporting

This section covers professional techniques used in aerospace and data science to make plots more informative and visually appealing.

### 1. Global Styling with `plt.style.use`
By setting a style at the beginning of your work, you ensure consistent, high-quality visuals. `'seaborn-v0_8-whitegrid'` is a favorite because it's clean and easy to read.

### 2. Twin Axes with `plt.twinx()`
When you have two variables with different units (like Kilometers for altitude and kg/m³ for density) but the same X-axis (Time), you use a twin axis. This allows you to overlay them without one 'squishing' the other.

### 3. Professional Saving
Using `bbox_inches='tight'` in `savefig` is a best practice. It automatically calculates the space needed for labels so they aren't cropped out.

```python
import matplotlib.pyplot as plt
import numpy as np

# Ensure the rocket data exists (in case previous cells weren't run)
class Rocket:
    def __init__(self):
        self.history = [0, 100, 200, 300, 400, 500, 450, 350, 200, 50]

if 'rocket' not in locals():
    rocket = Rocket()
if 'time_steps' not in locals():
    time_steps = np.arange(len(rocket.history)) * 0.1

# Apply professional styling
plt.style.use('seaborn-v0_8-whitegrid')

# Generate dummy density data (exponential decay as altitude increases)
density_history = [1.225 * np.exp(-h/8500) for h in rocket.history]

fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Altitude on the first Y-axis (left)
color1 = 'tab:red'
ax1.set_xlabel('Time (seconds)')
ax1.set_ylabel('Altitude (km)', color=color1)
ax1.plot(time_steps, [h/1000 for h in rocket.history], color=color1, linewidth=2, label='Altitude')
ax1.tick_params(axis='y', labelcolor=color1)

# Create a twin axis for Atmospheric Density (right)
ax2 = ax1.twinx()
color2 = 'tab:blue'
ax2.set_ylabel('Density (kg/m³)', color=color2)
ax2.plot(time_steps, density_history, color=color2, linewidth=2, linestyle='--', label='Density')
ax2.tick_params(axis='y', labelcolor=color2)

# Adding a Title
plt.title('Rocket Ascent: Altitude & Atmospheric Density')
fig.tight_layout()

# Saving properly - ensures no labels are cut off
plt.savefig('professional_rocket_plot.png', dpi=150, bbox_inches='tight')

plt.show()
```

## Matplotlib in ML, DL, and Quantum Research

### 1. Deep Learning: Training History
Visualizing loss and accuracy over epochs is the most common task in DL to detect overfitting or underfitting.

```python
import matplotlib.pyplot as plt
import numpy as np

# Simulated DL training data
epochs = np.arange(1, 21)
train_loss = np.exp(-epochs/5) + 0.1 * np.random.rand(20)
val_loss = np.exp(-epochs/6) + 0.2 + 0.1 * np.random.rand(20)

plt.figure(figsize=(8, 5))
plt.plot(epochs, train_loss, 'bo-', label='Training Loss')
plt.plot(epochs, val_loss, 'r--', label='Validation Loss')
plt.title('Deep Learning Model: Training vs Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Common ML Export practice: High DPI for papers/reports
plt.savefig('dl_training_curves.png', dpi=300, bbox_inches='tight')
plt.show()
```

### 2. Machine Learning: Model Evaluation
Confusion matrices and Heatmaps are essential for understanding classification performance.

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Simulated classes: [Rocket, Satellite, Debris]
y_true = [0, 1, 0, 2, 1, 1, 0, 2]
y_pred = [0, 1, 1, 2, 1, 0, 0, 2]
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Rocket', 'Satellite', 'Debris'],
            yticklabels=['Rocket', 'Satellite', 'Debris'])
plt.title('ML Classification Confusion Matrix')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')

plt.savefig('ml_confusion_matrix.pdf', bbox_inches='tight') # PDF for vector scaling
plt.show()
```

### 3. Quantum ML: State Visualizations
In Quantum ML, we often visualize probability distributions of bitstrings or Bloch spheres (using helper libraries, but rendered via Matplotlib).

```python
# Visualizing Quantum Probability Distributions
states = ['|00>', '|01>', '|10>', '|11>']
probabilities = [0.45, 0.05, 0.02, 0.48]

plt.figure(figsize=(7, 5))
plt.bar(states, probabilities, color='teal', alpha=0.8)
plt.axhline(y=0.5, color='red', linestyle='--', label='Max Entanglement Limit')
plt.title('QML: State Probability Distribution')
plt.ylabel('Probability')
plt.ylim(0, 1)
plt.legend()

plt.savefig('quantum_state_dist.png', transparent=True)
plt.show()
```
